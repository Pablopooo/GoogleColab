```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
for fn in uploaded.keys():
    name=fn
df = pd.read_csv(name, sep=",")
df.head(10)
```

```python
df.describe()
```

```python
#Distribucin de fuerza general de los Pokmon
plt.figure(figsize=(10, 5))
sns.histplot(df["Total"], bins=30, kde=True, color="green")
plt.title("Distribucin del Total de Fuerza")
plt.xlabel("Total")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["Speed"], y=df["Attack"], palette="dark", alpha=0.7)
plt.title("Relacin entre Velocidad y Ataque")
plt.xlabel("Velocidad")
plt.ylabel("Ataque")
plt.show()
```
